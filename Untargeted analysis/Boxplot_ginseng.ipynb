```r
library(ggplot2)
library(dplyr)
library(reshape)
```

```r
dat_r <- read.csv("Untargeted_root.csv", sep=",", stringsAsFactors = FALSE)
dat_b <- read.csv("Untargeted_berry.csv", sep=",", stringsAsFactors = FALSE)
```

```r
dat_r_norm <- dat_r
dat_r_norm[,3:73] <- data.frame(scale(dat_r[,3:73], center = TRUE, scale = TRUE))

dat_b_norm <- dat_b
dat_b_norm[,3:73] <- data.frame(scale(dat_b[,3:73], center = TRUE, scale = TRUE))
```

```r
dat_r_melt <- melt(data = dat_r_norm, id.vars = c("X", "Part"))
dat_b_melt <- melt(data = dat_b_norm, id.vars = c("X", "Part"))
```

```r
options(repr.plot.width=20, repr.plot.height=40)
ggplot(dat_r_melt, aes(x = variable, y=value)) + 
    geom_boxplot(outlier.shape = NA) +
    geom_jitter(width = 0.2, size = 2.5) +
    scale_x_discrete(limits = rev, position = 'top') +
    scale_y_reverse(limits = c(6.5,-2)) +
    theme_bw() +
    theme(axis.text.x = element_text(color='black', size = 25), axis.title.x = element_blank(), axis.text.y = element_text(color='black', size = 25), axis.title.y = element_blank()) +
    coord_flip()
ggsave("ginseng_root_box.png", width = 20, height = 40)
```

```r
options(repr.plot.width=20, repr.plot.height=40)
ggplot(dat_b_melt, aes(x = variable, y=value)) + 
    geom_boxplot(outlier.shape = NA) +
    geom_jitter(width = 0.2, size = 2.5) +
    scale_x_discrete(limits = rev) +
    scale_y_continuous(limits = c(-2,6.5)) +
    theme_bw() +
    theme(axis.text.x = element_text(color='black', size = 25), axis.title.x = element_blank(), axis.text.y = element_text(color='black', size = 25), axis.title.y = element_blank()) +
    coord_flip()
ggsave("ginseng_berry_box.png", width = 20, height = 40)
```
